# Visualize similarity

### Python, 12,708 full texts, 200 iterations

```python
from scipy.spatial.distance import cosine
def wrangle_sankey_data(X, Y, X_colors, p_thres=1e-6):
    label = list(X.columns) + ["Topic {}".format(y) for y in list(Y.columns)]
#     label = [" " for l in label]
    source, target, value, ps = [], [], [], []
    for i, x in enumerate(X.columns):
        for j, y in enumerate(Y.columns):
            j += len(X.columns)
            similarity, p = pearsonr(Y[y], X[x])
            source.append(i)
            target.append(j)
            if p < p_thres:
                value.append(similarity)
            else:
                value.append(0)
            ps.append(p)
    Y_colors = []
    for i, t in enumerate(target):
        idx = [id for id, val in enumerate(target) if val == t and ps[id] < p_thres]
        if len(idx) > 0:
            colors = [X_colors[source[id]] for id in idx]
            colors = [hex2rgb(hex) for hex in colors]
            values = [value[id] for id in idx]
            proportions = [val/sum(values) for val in values]
            rgb = [int(sum(np.multiply([c[i] for c in colors], proportions))) for i in range(3)]
            hex = rgb2hex(rgb[0], rgb[1], rgb[2])
            Y_colors.append(hex)
#             Y_colors.append(rgb)
        else:
            Y_colors.append("white")
    node_color = X_colors + Y_colors
    link_palette = ["#b4bbc4"]*5 + ["#d1b2ab"]*9 + ["#b3cebb"]*15 + ["#b6b3c6"]*10 + ["#ddd6af"]*3
    link_color = [link_palette[i] for i in source]
    return (label, source, target, value, node_color, link_color)

c = list(OrderedDict.fromkeys(rdoc["COLOR"]))
colors = [c[0]]*5 + [c[1]]*9 + [c[2]]*15 + [c[3]]*10 + [c[4]]*3
sankey_data = wrangle_sankey_data(llda, lda, colors)
```

```python
def plot_sankey(data, file_name):
    label, source, target, value, node_color, link_color = data
    data = dict(type="sankey", orientation="h",
                node = dict(pad=10, thickness=150,
                  line = dict(color="black", width=1.25),
                  label=label, color=node_color), 
                link = dict(source=source, target=target, 
                            value=value, color=link_color))
    layout = dict(font = dict(size=18, family="Avenir", color="black"), 
                  height=750, width=950, margin=dict(l=50,r=50,b=75,t=50,pad=50))
    fig = dict(data=[data], layout=layout)
    plotly.offline.plot(fig, filename=file_name, auto_open=False, show_link=False)
    return fig

sankey_fig = plot_sankey(sankey_data, "visuals/llda2lda.html")
py.iplot(sankey_fig, validate=False, show_link=False)
```
